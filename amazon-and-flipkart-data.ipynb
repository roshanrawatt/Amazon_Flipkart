# Combining the two datasets

```python

result = pd.merge(amazon, flip, how = "inner", on = 'product_name', left_index = False, 
         right_index = False, sort = True, suffixes = ("_x", "_y"), copy = True, indicator = False,
         validate=None)
```

```python
result.head()
```

```python
result.info()
```

# Instead of deleting all the columns from the dataset we will use filter function to retrieve the columns we need.

```python
result1 = result.filter(['product_name','retail_price_x', 'discounted_price_x', 'retail_price_y', 'discounted_price_y'], axis=1)
```

```python
result1.head(10)
```

```python
#Checking for null values
result1.isnull().sum()
```

```python
#Deleting the null values
result1 = result1.dropna(axis = 0)
```

# Creating the final dataset in accordance with the problem statement.

```python


final_df = pd.DataFrame().assign(product_name_in_flipkart = result1['product_name'], 
                                 retail_price_in_flipkart = result1['retail_price_x'], 
                                 discounted_price_in_flipkart = result1['discounted_price_x'],
                                 product_name_in_amazon = result1['product_name'],
                                 retail_price_in_amazon = result1['retail_price_y'],
                                 discounted_price_in_amazon = result1['discounted_price_y'])
```
